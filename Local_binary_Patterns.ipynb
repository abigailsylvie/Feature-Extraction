```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from sklearn.datasets import load_sample_images

# Step 1: Load a sample image and convert to grayscale
data = load_sample_images()
img = data.images[0]
gray_img = rgb2gray(img)

# Step 2: Define the LBP function
def lbp(new_image):
    arr = np.zeros(new_image.shape)
    for i in range(1, new_image.shape[0]-1):
        for j in range(1, new_image.shape[1]-1):
            n = new_image[i][j]
            n1 = new_image[i][j-1] > n
            n2 = new_image[i-1][j-1] > n
            n3 = new_image[i-1][j] > n
            n4 = new_image[i-1][j+1] > n
            n5 = new_image[i][j+1] > n
            n6 = new_image[i+1][j+1] > n
            n7 = new_image[i+1][j] > n
            n8 = new_image[i+1][j-1] > n
            result = (n1 * 2**0 + n2 * 2**1 + n3 * 2**2 + n4 * 2**3 +
                      n5 * 2**4 + n6 * 2**5 + n7 * 2**6 + n8 * 2**7)
            arr[i][j] = result
    return arr

# Step 3: Apply LBP and visualize the result
lbp_result = lbp(gray_img)

plt.imshow(lbp_result, cmap='gray')
plt.title('Local Binary Pattern (LBP)')
plt.axis('off')
plt.show()
```

```python
# Step 1: Import the modules
from sklearn.datasets import load_iris
import numpy as np
from matplotlib import pyplot as plt
from numpy import linalg

# Step 2: Load and prepare the data
iris = load_iris()
Data = iris.data  # shape (150, 4)

# Step 3: Apply PCA manually
mean = np.mean(Data, axis=0)
centered_data = Data - mean
cov_matrix = np.matmul(centered_data.T, centered_data)
val, vec = linalg.eig(cov_matrix)
Data_transformed = np.matmul(centered_data, vec)

print(Data_transformed.shape)

# Step 4: Visualize PCA-transformed features for class 1 and 2
Data1 = Data_transformed[50:100, :]   # Class 1: Versicolor
Data2 = Data_transformed[100:150, :]  # Class 2: Virginica
index = np.arange(0, 50)  # 50 samples each class

plt.plot(index, Data1[:, 0], 'rs', label='Class 1 (Versicolor)')
plt.plot(index, Data2[:, 0], 'bs', label='Class 2 (Virginica)')
plt.xlabel('Sample Index')
plt.ylabel('PCA Feature 1')
plt.title('PCA Feature 1 Comparison')
plt.legend()
plt.grid(True)
plt.show()

plt.plot(index, Data1[:, 1], 'rs', label='Class 1 (Versicolor)')
plt.plot(index, Data2[:, 1], 'bs', label='Class 2 (Virginica)')
plt.xlabel('Sample Index')
plt.ylabel('PCA Feature 2')
plt.title('PCA Feature 2 Comparison')
plt.legend()
plt.grid(True)
plt.show()

```

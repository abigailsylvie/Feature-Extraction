```python
import pywt
import cv2
from matplotlib import pyplot as plt

# Load the image
img = plt.imread('pexels-frans-van-heerden-201846-1463295.jpg')

# Convert to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Haar wavelet transform
coef = pywt.dwt2(gray_img, 'haar')
LL, (LH, HL, HH) = coef

# Plot the results
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(LL, cmap='gray')
plt.title('Approximation (LL)')

plt.subplot(2, 2, 2)
plt.imshow(LH, cmap='gray')
plt.title('Horizontal Details (LH)')

plt.subplot(2, 2, 3)
plt.imshow(HL, cmap='gray')
plt.title('Vertical Details (HL)')

plt.subplot(2, 2, 4)
plt.imshow(HH, cmap='gray')
plt.title('Diagonal Details (HH)')

plt.tight_layout()
plt.show()
```
